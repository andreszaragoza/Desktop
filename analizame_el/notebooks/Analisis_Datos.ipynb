### Normalizar Columnas Anidadas

```python
def extract_names(json_string, key='name'):
    try:
        return ", ".join([item[key] for item in ast.literal_eval(json_string)])
    except (ValueError, KeyError, TypeError):
        return None

df['platforms'] = data['platforms'].apply(lambda x: extract_names(x, key='platform'))
df['parent_platforms'] = data['parent_platforms'].apply(lambda x: extract_names(x, key='platform'))
df['genres'] = data['genres'].apply(extract_names)  
```
