# Analisis Estadistico

### Análisis ANOVA o t-test para ver si ciertos géneros o etiquetas influyen en las calificaciones.

```python
df['genres_list'] = df['genres'].apply(lambda x: [g['name'] for g in eval(x)] if pd.notna(x) else [])

df_expanded = df.explode('genres_list').dropna(subset=['genres_list'])
df_expanded = df_expanded[['name', 'genres_list', 'rating']].dropna()

'''Agrupar por géneros y preparar los datos para ANOVA'''
groups = {genre: df_expanded[df_expanded['genres_list'] == genre]['rating'].values for genre in df_expanded['genres_list'].unique()}

anova = stats.f_oneway(*[ratings for ratings in groups.values() if len(ratings) > 5])

nova = {
    "p-value": anova.pvalue,
    "F-statistic": anova.statistic,
    "Significativo": "Sí" if anova.pvalue < 0.05 else "No"
}

top_genres = df_expanded['genres_list'].value_counts().head(10).to_dict()

nova, top_genres
```

### **Resultados del ANOVA**

- **F-statistic:** 26.58  
- **p-value:** \(9.41 \times 10^{-89}\) (prácticamente 0)
---
### **Conclusión:**
La prueba ANOVA muestra que **hay diferencias estadísticamente significativas en las calificaciones (`rating`) según el género**. Esto indica que, al menos, un grupo de géneros tiene una media de calificación distinta de los demás.

```python
df['released'] = pd.to_datetime(df['released'], errors='coerce')
df['release_year'] = df['released'].dt.year


import ast
from scripts.data_cleaning import extract_main_genre
df['main_genre'] = df['genres'].apply(extract_main_genre)

''' Calcular la mediana de metacritic agrupando por 'release_year' y 'main_genre'''
median_metacritic = df.groupby(['release_year', 'main_genre'])['metacritic'].median()

''' Imputar los valores faltantes de 'metacritic' con la mediana correspondiente'''
def impute_metacritic(row):
    if pd.isna(row['metacritic']):
        return median_metacritic.get((row['release_year'], row['main_genre']), df['metacritic'].median())
    return row['metacritic']

df['metacritic'] = df.apply(impute_metacritic, axis=1)
'''missing faltantes'''
missing_metacritic_after = df['metacritic'].isnull().sum()
missing_metacritic_after

global_median_metacritic = df['metacritic'].median()
df['metacritic'].fillna(global_median_metacritic, inplace=True)

missing_metacritic_final = df['metacritic'].isnull().sum()
missing_metacritic_final
```

### Resumen:
- La cantidad de valores faltantes en la columna metacritic ha disminuido de 3328 a 147 después de la imputación utilizando la mediana por año de lanzamiento y género principal.
- Se imputaron los valores utilizando la mediana por género principal y año de lanzamiento.
- Los valores restantes se completaron con la mediana global.

```python
def extract_store_names(store_str):
    try:
        store_list = ast.literal_eval(store_str)
        if isinstance(store_list, list):
            return [store['store']['name'] for store in store_list if 'store' in store]
    except (ValueError, SyntaxError):
        return []
    return []


df['store_list'] = df['stores'].apply(extract_store_names)

def get_top_games_by_store(store_name, top_n=5):
    store_games = df[df['store_list'].apply(lambda x: store_name in x)]
    top_games = store_games.sort_values(by='metacritic', ascending=False).head(top_n)
    return top_games[['name', 'metacritic', 'release_year', 'main_genre', 'store_list']]

top_steam_games = get_top_games_by_store('Steam')
top_playstation_games = get_top_games_by_store('PlayStation Store')
top_nintendo_games = get_top_games_by_store('Nintendo Store')
top_xbox_games = get_top_games_by_store('Xbox Store')
top_steam_games.head(), top_playstation_games.head(), top_nintendo_games.head(), top_xbox_games.head()
```

```python
def plot_top_games(store_name, top_games_df):
    plt.figure(figsize=(10, 6))
    plt.barh(top_games_df['name'], top_games_df['metacritic'], color='skyblue')
    plt.xlabel('Puntaje Metacritic')
    plt.title(f'Mejores Juegos en {store_name}')
    plt.gca().invert_yaxis()
    plt.show()

plot_top_games("Steam", top_steam_games)
plot_top_games("PlayStation Store", top_playstation_games)
plot_top_games("Nintendo Store", top_nintendo_games)
plot_top_games("Xbox Store", top_xbox_games)
```
