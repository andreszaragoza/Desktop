# Analisis Estadistico

### Análisis ANOVA o t-test para ver si ciertos géneros o etiquetas influyen en las calificaciones.

```python
df['genres_list'] = df['genres'].apply(lambda x: [g['name'] for g in eval(x)] if pd.notna(x) else [])

df_expanded = df.explode('genres_list').dropna(subset=['genres_list'])
df_expanded = df_expanded[['name', 'genres_list', 'rating']].dropna()

'''Agrupar por géneros y preparar los datos para ANOVA'''
groups = {genre: df_expanded[df_expanded['genres_list'] == genre]['rating'].values for genre in df_expanded['genres_list'].unique()}

anova = stats.f_oneway(*[ratings for ratings in groups.values() if len(ratings) > 5])

nova = {
    "p-value": anova.pvalue,
    "F-statistic": anova.statistic,
    "Significativo": "Sí" if anova.pvalue < 0.05 else "No"
}

top_genres = df_expanded['genres_list'].value_counts().head(10).to_dict()

nova, top_genres
```

### **Resultados del ANOVA**

- **F-statistic:** 26.58  
- **p-value:** \(9.41 \times 10^{-89}\) (prácticamente 0)
---
### **Conclusión:**
La prueba ANOVA muestra que **hay diferencias estadísticamente significativas en las calificaciones (`rating`) según el género**. Esto indica que, al menos, un grupo de géneros tiene una media de calificación distinta de los demás.

- Promedios por rango de tiempo jugado:

- Juegos con menor tiempo jugado (0-1h) tienen una calificación promedio baja (~1.70).
- Juegos en el rango 5-10h tienen la calificación promedio más alta (~2.98).

## Tendencias de popularidad según la fecha de lanzamiento (released).

```python
df['released'] = pd.to_datetime(df['released'], errors='coerce')
games_data_valid = df.dropna(subset=['released'])
games_data_valid['release_year'] = games_data_valid['released'].dt.year

popularity_by_year = games_data_valid.groupby('release_year').size().reset_index(name='count')

plt.figure(figsize=(12, 6))
plt.plot(popularity_by_year['release_year'], popularity_by_year['count'], marker='o')
plt.title('Tendencias de Popularidad Según el Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Número de Juegos Lanzados')
plt.grid(True)
plt.show()
```

##### Aquí tienes un gráfico que muestra las tendencias de popularidad según el año de lanzamiento:

- El eje X representa el año de lanzamiento.
- El eje Y muestra el número de juegos lanzados en ese año.
- Puedes observar cómo ha cambiado la cantidad de juegos lanzados a lo largo del tiempo, con un crecimiento notable hasta cierto punto y luego una disminución.

```python
expanded_genres_data = games_data_valid.explode('genres_list')

popularity_by_genre_year = expanded_genres_data.groupby(['release_year', 'genres_list']).size().reset_index(name='count')

popularity_pivot = popularity_by_genre_year.pivot(index='release_year', columns='genres_list', values='count').fillna(0)

top_genres = popularity_by_genre_year.groupby('genres_list')['count'].sum().nlargest(5).index
popularity_top_genres = popularity_pivot[top_genres]

plt.figure(figsize=(14, 8))
for genre in top_genres:
    plt.plot(popularity_top_genres.index, popularity_top_genres[genre], marker='o', label=genre)

plt.title('Tendencias de Popularidad por Género y Año')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Número de Juegos Lanzados')
plt.legend(title='Géneros', loc='upper left')
plt.grid(True)
plt.show()
```

##### Aquí tienes un gráfico que muestra las tendencias de popularidad por género y año:

- Cada línea representa uno de los 5 géneros más populares basados en el número total de lanzamientos.
- El eje X muestra los años de lanzamiento.
- El eje Y representa el número de juegos lanzados en ese género en cada año.

## Análisis de (genres) o (tags) más comunes entre los juegos mejor calificados.
